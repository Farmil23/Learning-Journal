```python
from google_play_scraper import reviews
import pandas as pd

app_id = 'com.instagram.android'  # Package name Instagram
count = 40000  # Jumlah ulasan yang ingin diambil
all_reviews = []
continuation_token = None

while len(all_reviews) < count:
    result, continuation_token = reviews(
        app_id,
        lang='en',
        country='us',
        count=1000,
        continuation_token=continuation_token
    )
    all_reviews.extend(result)
    print(f"Total reviews collected: {len(all_reviews)}")

    if not continuation_token:
        break

# Simpan ke DataFrame
df = pd.DataFrame(all_reviews)
df.to_csv('instagram_reviews.csv', index=False)
print("Selesai! File disimpan sebagai instagram_reviews.csv")
```
